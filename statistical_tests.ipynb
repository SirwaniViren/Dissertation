```python
# Hamming Distance
distance = []
for disease in list(org_df):
    d = np.sum(org_df[disease].to_list() != synth_df[disease].to_list())
    distance.append(d)

# Calculate the average distance
distance = np.mean(distance)

# Print the result
print(f"Hamming distance: {distance:.2f}")
```

```python
from sklearn.metrics import jaccard_score

similarity = []
for disease in list(org_df.iloc[:, 0:30]):
    s = jaccard_score(org_df[disease].to_list(), synth_df[disease].to_list())
    similarity.append(s)


# Calculate the overall similarity
similarity = np.mean(similarity)

# Print the result
print(f"Jaccard similarity: {similarity:.2f}")
```

```python
from scipy.stats import chi2_contingency

p_val_list = []
chi2_stat_list = []

for disease in list(org_df.iloc[:, 0:30]):

    mean1 = np.mean(list(org_df[disease]))
    mean2 = np.mean(list(synth_df[disease]))
    std1 = np.std(list(org_df[disease]))
    std2 = np.std(list(synth_df[disease]))

    contingency_table = np.array([[list(org_df[disease]).count(1), list(org_df[disease]).count(0)],
                                [list(synth_df[disease]).count(1), list(synth_df[disease]).count(0)]])

    chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)
    p_val_list += p_val
    chi2_stat_list += chi2_stat
    
print(f"Chi-Squared Statistic: {np.mean(chi2_stat_list)}\np-value: {np.mean(p_val_list)}")
```

```python
contingency_table = np.array([org_df.iloc[:, 0:30].sum().values, synth_df.iloc[:, 0:30].sum().values])
chi2_stat, p_val, dof, ex = chi2_contingency(contingency_table)
print(f"Chi-Squared Statistic: {chi2_stat}\nDegrees of Freedom: {dof}\np-value: {p_val}")
```
